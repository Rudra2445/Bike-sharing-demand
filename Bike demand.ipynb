```python
#Assigning string values to different season instead of numeric values
# 1 = spring
dataset.loc[(dataset['season'] == 1),'season'] = 'spring'

# 2 = summer 
dataset.loc[(dataset['season'] == 2), 'season'] = 'summer'

#3 = fall
dataset.loc[(dataset['season'] == 3), 'season'] = 'fall'

#4 = winter
dataset.loc[(dataset['season'] == 4), 'season'] = 'winter'
```

```python
#Assigning string values to different months instead of numeric values
def object_map_mnths(x):
    return x.map({1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sept',10:'Oct',11:'Nov',12:'Dec'})
```

```python
dataset[['mnth']] = dataset[['mnth']].apply(object_map_mnths)
```

```python
dataset['mnth'].astype('category').value_counts()
```

```python
dataset['holiday'].astype('category').value_counts()
```

```python
def str_map_weekday(x):
    return x.map({1:'Mon', 2:'Tue', 3:'Wed', 4:'Thur',5:'Fri',6:'Sat',0:'Sun'})
```

```python
dataset[['weekday']] = dataset[['weekday']].apply(str_map_weekday)
```

```python
dataset['weekday'].astype('category').value_counts()
```

```python
dataset['workingday'].astype('category').value_counts()
```

```python
# 1 = clear/partly cloudy
dataset.loc[(dataset['weathersit'] == 1),'weathersit'] = 'A'

# 2 = mist/more cloudy
dataset.loc[(dataset['weathersit'] == 2),'weathersit'] = 'B'

# 3 = light snow / heavy rain
dataset.loc[(dataset['weathersit'] == 3),'weathersit'] = 'C'
```

## Data Preparation

```python
dataset_categorical = dataset.select_dtypes(include='object')
```

```python
dataset_categorical.head()
```

```python
dataset_dummies = pd.get_dummies(dataset_categorical, drop_first= True,).astype(int)
dataset_dummies.head()
```

```python
# Dropping categorical Variable columns
dataset = dataset.drop(list(dataset_categorical.columns),axis=1)
```

```python
dataset
```

```python
# Concatenate dummy variables with the dataset
dataset = pd.concat([dataset,dataset_dummies],axis=1)
dataset.head()
```

```python
dataset = dataset.drop(['instant','dteday'], axis = 1, inplace = False)
dataset.head()
```

```python
# There are to many parameters so we are removing some of them 
# By using a module RFE(Reccursive Feature Elimination)
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Create a linear regression model
lm = LinearRegression()

# Create an RFE object with the linear regression model and number of features to select
#estimator parameter specifies the machine learning model that will be used for feature selection.
# n_features_to_select parameter specifies the number of features to select after performing RFE.
rfe = RFE(estimator=lm, n_features_to_select=15)

# Fitting RFE to our data
rfe.fit(X_train, Y_train)

print("Selected Features:", rfe.support_)
print("Feature Rankings:", rfe.ranking_)
```

```python
col1 = X_train.columns[rfe.support_]
print("Selected Features:", col1)
```

```python
X_train_rfe = X_train[col1]

X_train_lm = sm.add_constant(X_train_rfe)
lm1 = sm.OLS(Y_train, X_train_rfe).fit()
lm1.summary()
```

```python
X_train_rfe
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
a = X_train_rfe
```

```python
# Evaluating the VIF's 
vif = pd.DataFrame()
vif['features'] = a.columns
vif['VIF'] = [variance_inflation_factor(a.values , i) for i in range(a.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF',ascending = False)
vif
```

```python
lm = LinearRegression()

rfe2 = RFE(estimator=lm, n_features_to_select=7)

rfe2.fit(X_train, Y_train)

print("Selected Features:", rfe2.support_)
print("Feature Rankings:", rfe2.ranking_)
```

```python
X_train
```

```python
col2 = X_train.columns[rfe2.support_]
print("Selected Features:", col2)

X_train_rfe2 = X_train[col2]

X_train_rfe2 = sm.add_constant(X_train_rfe2)
lm2 = sm.OLS(Y_train, X_train_rfe2).fit()
lm2.summary()
```

```python
b = X_train_rfe2.drop('const',axis =1)

# Evaluating the VIF's 
vif1 = pd.DataFrame()
vif1['features'] = b.columns
vif1['VIF'] = [variance_inflation_factor(b.values , i) for i in range(b.shape[1])]
vif1['VIF'] = round(vif1['VIF'], 2)
vif1 = vif1.sort_values(by = 'VIF',ascending = False)
vif1
```
